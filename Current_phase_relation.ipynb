```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
#двухконтактный SQUID 

F1 = np.linspace(0, 3, 500)
I = []

for i in range(0,len(F1)):
    It = np.abs(np.cos(np.pi*F1[i]))
    I.append(It)
    
fig, ax = plt.subplots(figsize=(10,6)) 
ax.plot(F1, I)
plt.ylabel("$I_{max}/2I_{c}$", fontsize=14)
plt.xlabel("$\Phi/\Phi_{0}$", fontsize=14)
# ax.set_ylabel('$I_{max}/I_{c}$')
# ax.set_xlabel('$\Phi/\Phi_{0}$')
plt.show()
```

```python
#двухконтактный SQUID с ассиметричными SIS

F1 = np.linspace(0, 3, 1000) #поток через пелю
I = [] #полный ток
I2 = []

alpha = 1.5 # соотношение криттоков левого и правого контакта
alpha2 = 5

phi2 = np.linspace(0, np.pi, 500)

for i in range(0,len(F1)):
    Current = []
    for j in range(0, len(phi2)):
        # J = 2* np.cos(np.pi*F1[i])*np.sin(phi2[j]+np.pi*F1[i]) # верно для симметричного сквида
        J = (np.sin(phi2[j]+2*np.pi*F1[i]) + alpha*np.sin(phi2[j]))
        Current.append(J)
    I.append(max(Current))

for i in range(0,len(F1)):
    Current = []
    for j in range(0, len(phi2)):
        # J = 2* np.cos(np.pi*F1[i])*np.sin(phi2[j]+np.pi*F1[i]) # верно для симметричного сквида
        J = (np.sin(phi2[j]+2*np.pi*F1[i]) + alpha2*np.sin(phi2[j]))
        Current.append(J)
    I2.append(max(Current))

I_0 = [] #полный ток
for i in range(0,len(F1)):
    It = np.abs(np.cos(np.pi*F1[i]))
    I_0.append(It)

Norm = max(I)
Norm_0 = max(I_0)
Norm2= max(I2)
fig, ax = plt.subplots(figsize=(10,6)) 
ax.plot(F1, I_0/Norm_0, label='Symmetric SIS', color = 'mediumorchid')
ax.plot(F1, I/Norm, label = r'Asymmetric SIS, $\alpha$ = 1.5', color = 'royalblue')
ax.plot(F1, I2/Norm2, label = r'Asymmetric SIS, $\alpha$ = 5', color = 'darkturquoise')
ax.legend(fontsize=14, loc='lower right')
plt.ylabel("$I_{max}/I_{0}$", fontsize=14)
plt.xlabel("$\Phi/\Phi_{0}$", fontsize=14)
plt.show()
```

```python
# двухконтактный SQUID с пи-контактом и без

F1 = np.linspace(0, 3, 1000) #поток через пелю
I_0 = [] #полный ток
I_pi = []


phi2 = np.linspace(-np.pi, np.pi, 1000)

for i in range(0,len(F1)):
    Current = []
    for j in range(0, len(phi2)):
        J = (-np.sin(phi2[j]+2*np.pi*F1[i]) + np.sin(phi2[j]))
        Current.append(J)
    I_pi.append(max(Current))

for i in range(0,len(F1)):
    It = np.abs(np.cos(np.pi*F1[i]))
    I_0.append(It)
    

Norm_pi = max(I_pi)
Norm_0 = max(I_0)
fig, ax = plt.subplots(figsize=(10,6)) 
ax.plot(F1, I_pi/Norm_pi, label='SIS + SIS', color = 'royalblue')
ax.plot(F1, I_0/Norm_0, label=r'SIS + $\pi$ - junction', color = 'darkorange')
plt.ylabel("$I_{max}/I_{0}$", fontsize=14)
plt.xlabel("$\Phi/\Phi_{0}$", fontsize=14)
ax.legend(fontsize=14, loc='lower right')
plt.show()
```

```python
# симметричных трехконтактный SQUID

F1 = np.linspace(0, 3, 1000) #поток через пелю
I = [] #полный ток

phi2 = np.linspace(-np.pi/2, np.pi/2, 1000)

for i in range(0,len(F1)):
    Current = []
    for j in range(0, len(phi2)):
        J = (np.sin(phi2[j]+2*np.pi*F1[i]) + np.sin(phi2[j])+np.sin(phi2[j]-2*np.pi*F1[i]))
        Current.append(J)
    I.append(max(Current))

Norm = max(I)
fig, ax = plt.subplots(figsize=(10,6)) 
ax.plot(F1, I/Norm)
plt.ylabel("$I_{max}/2I_{c}$", fontsize=14)
plt.xlabel("$\Phi/\Phi_{0}$", fontsize=14)
plt.show()
```

```python
# симметричных трехконтактный SQUID с пи-контактом посередине

F1 = np.linspace(0, 3, 1000) #поток через пелю
I = [] #полный ток

phi2 = np.linspace(-np.pi/2, np.pi/2, 1000)

for i in range(0,len(F1)):
    Current = []
    for j in range(0, len(phi2)):
        J = (np.sin(phi2[j]+2*np.pi*F1[i]) - np.sin(phi2[j])+np.sin(phi2[j]-2*np.pi*F1[i]))
        Current.append(J)
    I.append(max(Current))

Norm = max(I)
fig, ax = plt.subplots(figsize=(10,6)) 
ax.plot(F1, I/Norm)
plt.ylabel("$I_{max}/2I_{c}$", fontsize=14)
plt.xlabel("$\Phi/\Phi_{0}$", fontsize=14)
plt.show()
```
